```python
# 네이버에서 검색어 입력받아 검색 한 후 블로그 메뉴를 선택하고
# 오른쪽에 있는 검색옵션 버튼을 눌러서
# 정렬 방식과 기간을 입력하기

#Step 0. 필요한 모듈과 라이브러리를 로딩합니다.
import sys # 시스템
import os # 시스템
import pandas as pd # 데이터 분석
import numpy as np # 숫자

from bs4 import BeautifulSoup # 데이터 파싱
from selenium import webdriver # 브라우저 자동화
import time # 서버응답시간 대기
from tqdm import tqdm_notebook # for문 진행상황 시각화
```

```python
query_txt="건강식품"
start_date="20190101"
end_date="20191231"
```

# 크롤링할 뉴스 url 수집하기

```python
#Step 1. 크롬 웹브라우저 실행
path = "chromedriver.exe"  # 맥 기준  # 윈도우는 "chromedriver.exe"

driver = webdriver.Chrome(path)
# 사이트 주소는 네이버
driver.get('http://www.naver.com')
time.sleep(2)

#Step 2. 네이버 검색창에 "검색어" 검색
element = driver.find_element_by_id("query")
element.send_keys(query_txt)  # query_txt는 위에서 입력한 키워드
element.submit()
time.sleep(1)

# '검색옵션' 클릭
driver.find_element_by_link_text("검색옵션").click( )

# '뉴스' 클릭
driver.find_element_by_link_text("뉴스").click( )
time.sleep(1)

# Step 3. 정렬 : "관련도순" 
# 개발자 도구에서 정렬 버튼의 id 가 보이지 않습니다.
# 이럴 경우 쉽게 사용할 수 있는 방법이 바로 xpath 를 이용하는 방법입니다.
# xpath는 개발자 도구에서 해당 메뉴 부분을 마우스 오른쪽 버튼을 누르고 
# copy -> copy xpath 를 선택하면 됩니다
driver.find_element_by_xpath("""//*[@id="snb"]/div/ul/li[1]/a""").click( )  # 정렬 버튼의 xpath 클릭
driver.find_element_by_xpath("""//*[@id="snb"]/div/ul/li[1]/div/ul/li[1]/a""").click( ) # 관련도순 xpath
time.sleep(1)

# Step 4. 날짜 입력
driver.find_element_by_xpath("""//*[@id="snb"]/div/ul/li[2]/a""").click( )
time.sleep(2)

# 이 부분이 아주 중요합니다.
# 날짜 부분에 날짜를 입력할 때 입력 속도가 너무 빠를 경우 날짜가 입력이 되다가 
# 오타가 나오는 경우가 많습니다.
# 그래서 날짜를 입력할 때 for 반복문을 사용해서 1 글자씩 입력하도록 코딩했습니다.

# 시작 날짜 입력하기
s_date = driver.find_element_by_id("news_input_period_begin")
driver.find_element_by_id("news_input_period_begin").click()
s_date.clear( )  # 날짜 입력 부분에 기존에 입력되어 있던 날짜를 제거합니다. 
time.sleep(1)
# 아래 코드가 날짜를 for 반복문으로 1 글자씩 입력하는 부분입니다.
for c in start_date: # 데이터값이 뒤죽박죽 들어갈 수 있어 for문 + sleep 활용
    s_date.send_keys(c)
    time.sleep(0.1)
time.sleep(1)

# 종료 날짜 입력하기
e_date = driver.find_element_by_id("news_input_period_end")
driver.find_element_by_id("news_input_period_end").click()
e_date.clear()
time.sleep(1)

for c in end_date:
    e_date.send_keys(c)
    time.sleep(0.1)
    
# Step5. 날짜 입력 "적용" 버튼을 클릭 합니다.  
driver.find_element_by_class_name("tx").click()
time.sleep(3)

# '연합뉴스' 언론사만 선택
driver.find_element_by_xpath("""//*[@id="snb"]/div/ul/li[5]/a""").click( ) # 언론사
driver.find_element_by_xpath("""//*[@id="order_cat"]/div[1]/div/a[2]""").click( ) # 방송/통신
driver.find_element_by_xpath("""//*[@id="ca_1001"]""").click( ) # 연합뉴스
driver.find_element_by_xpath("""//*[@id="snb"]/div/ul/li[5]/div/span/span[1]/button/span""").click( ) # 확인

# # Step 8. 상세 검색버튼을 클릭 후 제외할 단어들을 설정합니다.
# ele2 = driver.find_element_by_id("inpop3")
# ele2.send_keys(query_txt2)
# ele2.send_keys(',')
# ele2.send_keys(query_txt3)
# driver.find_element_by_css_selector(".btn_ft.ty_green._search").click( )

## 글 url 크롤링 시작
```

```python
import math

# 크롤링 page num 구하기

article_num = 20 # 크롤링 원하는 기사 갯수
page_num = math.ceil(article_num/10)
print(page_num,'page')
```

```python
url_list = []
title_list = []

for i in range(1, page_num+1):
    
    # URL 크롤링 시작
    articles = ".news_tit"
    article_raw = driver.find_elements_by_css_selector(articles)

    # url 크롤링 시작    
    for article in article_raw:
        url = article.get_attribute('href')   
        url_list.append(url)
    time.sleep(1)

    # 제목 크롤링 시작    
    for article in article_raw:
        title = article.text
        title_list.append(title)

        print(title)
        
    driver.find_element_by_xpath("""//*[@id="main_pack"]/div[3]/div/a[2]/i""").click( )

print("")
print('url갯수: ', len(url_list))
print('title갯수: ', len(title_list))
```

```python
df = pd.DataFrame({'url':url_list, 'title':title_list})
df
```

```python
#  엑셀로 저장하기
df.to_excel("health_article urllist_2019.xlsx")
```

# 뉴스 내용 크롤링하기

```python
import sys
import os
import pandas as pd
import numpy as np
```

```python
# "url_list.csv" 불러오기
url_load = pd.read_excel("health_article urllist_2019.xlsx") # 기본 모델

num_list = len(url_load)

print(num_list)
url_load
```

```python
dict = {}  # 전체 크롤링 데이터를 담을 그릇

# ★수집할 글 갯수 정하기
number = num_list
for i in tqdm_notebook(range(0, number)):
    # 글 띄우기
    url = url_load['url'][i]
    driver = webdriver.Chrome("chromedriver.exe")  # 윈도우는 "chromedriver.exe"
    driver.get(url)   # 글 띄우기
    
    # 크롤링
    
    try : 
        target_info = {}

        # 제목 크롤링 시작
        overlays = ".tit"                                 
        tit = driver.find_element_by_css_selector(overlays)          # title
        title = tit.text

        # 언론사 
        press = '연합뉴스'

        # 날짜 크롤링
        overlays = ".update-time"                                 
        date = driver.find_element_by_css_selector(overlays)         # datetime
        datetime = date.text

        # 내용 크롤링
        overlays = "div.story-news.article > p"                                 
        contents = driver.find_elements_by_css_selector(overlays)    

        content_list = []
        for content in contents:
            content_list.append(content.text)           
 
        content_str = ' '.join(content_list)                         # content_str

        # 글 하나는 target_info라는 딕셔너리에 담기게 되고,
        target_info['datetime'] = datetime
        target_info['press'] = press
        target_info['title'] = title
        target_info['content'] = content_str

        # 각각의 글은 dict라는 딕셔너리에 담기게 됩니다.
        dict[i] = target_info
        time.sleep(1)
        
        # 크롤링이 성공하면 글 제목을 출력하게 되고,
        print(i, title)

        # 글 하나 크롤링 후 크롬 창을 닫습니다.
        driver.close()       
    
    # 에러나면 현재 크롬창을 닫고 다음 글(i+1)로 이동합니다.
    except:
        driver.close()
        time.sleep(1)
        continue
    
    # 중간,중간에 파일로 저장하기
    if i % 100 == 0:
        # 판다스로 만들기
        import pandas as pd
        result_df = pd.DataFrame.from_dict(dict, 'index')

        # 저장하기
        result_df.to_excel("health_article_2019.xlsx")
        time.sleep(3)

print('수집한 글 갯수: ', len(dict))
print(dict)
```
